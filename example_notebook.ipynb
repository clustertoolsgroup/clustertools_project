```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from clustertools.load_data import load_fuzzy_data
```

```python
x = load_fuzzy_data()
```

```python
x
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1],alpha=.3)
```

```python
from clustertools.models.distance import KMeans
```

```python
kmeans_obj =  KMeans(x,5)
```

```python
kmeans_obj.fit()
```

```python
labels = kmeans_obj.cluster_labels
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1],alpha=.3,c =labels)
```

```python
from clustertools.models.distance import Regspace
```

```python
reg = Regspace(x,10,.3)
```

```python
labels = reg.cluster_labels
```

```python
plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1],alpha=.3,c =labels)
```

```python
from clustertools.models.graph import SpectralClustering
```

```python
spc = SpectralClustering(x,5,similarity_measure='gaussian',bandwidth=.3,laplacian='normalized')
```

```python
spc.fit()
```

```python
labels = spc.cluster_labels
plt.figure(figsize=(10,10))
plt.scatter(x[:,0],x[:,1],alpha=.3,c =labels)
```
