```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
from clustertools.load_data import load_fuzzy_data
```

```python
x = load_fuzzy_data()
```

```python
from importlib import reload
from clustertools.models import similarity
similarity = reload(similarity)
sample = 2
```

```python
aff = similarity.AffinityPropagation(damp=.4,max_iter= 1000, data=x[::sample,:],sensitivity_weights='min',n_break_storage=10)
```

```python
aff.fit()
```

```python
labels = aff.cluster_labels
plt.scatter(x[::sample,0],x[::sample,1],c=labels,alpha = .4)
```
